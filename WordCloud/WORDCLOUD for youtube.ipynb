```python
import numpy as np
import matplotlib.pyplot as plt
from wordcloud import WordCloud

# Read the text from document ‘sample_text.txt’
text = open('astropi.txt', encoding="utf8").read()

# makes the circle using numpy
x, y = np.ogrid[:300, :300]
mask = (x - 150) ** 2 + (y - 150) ** 2 > 130 ** 2
mask = 255 * mask.astype(int)

# Generate a word cloud of document
wordcloud = WordCloud(background_color="white",
 mask=mask, contour_width=0.1, contour_color="black")
wordcloud.generate(text)

#Display the generated image:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
from os import path
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS

# Read the text from document ‘sample_text.txt’
text = open('astropi.txt', encoding="utf8").read()
# read the mask image from bottle
mask = np.array(Image.open("e.png"))
#plt.imshow(bottle_mask)
stopwords = set(STOPWORDS)

wordcloud = WordCloud(background_color="white", mask=mask,
               stopwords=stopwords, contour_width=3)

# generate word cloud
wordcloud.generate(text)

# Display the generated image:
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```
